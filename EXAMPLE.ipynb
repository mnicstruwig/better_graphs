## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
from better_graphs import better_graphs
from better_graphs import defaults
```

## Generate some data

```python
points = np.random.rand(100, 2)*100
more_points = np.random.rand(50, 2)*50+25

x = points[:, 0]
x2 = more_points[:, 0]

y = points[:, 1]
y2 = more_points[:, 1]
```

## Before

```python
plt.plot(x, y, 'o')
plt.plot(x2, y2, 'o')
```

# Let's make an attempt at improving things
## We'll first set some sane defaults (courtesy of Seaborn and some matplotlib tweaks)

```python
defaults.set_pretty_defaults(font='CMU Serif')
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(x, y, 'o')
plt.plot(x2, y2, 'o')
```

## Let's first try the basic functionality

```python
better_graphs.make_better_graph(ax)
fig
```

## Taking things further: customizing tick labels
We already have a graph that is publication-ready, but let's try improve things further.

```python
better_graphs.make_better_graph(ax, x_new_ticks={100:'100m'}, y_new_ticks={0:'0\%', 100:'100\%'})
fig
```

## We can make entirely custom axes too...

```python
better_graphs.make_better_graph(ax, x_new_ticks={0:0, 50:50, 100:'100m'}, only_new_ticks='x')
fig
```

## This allows you to remove as much chart junk as you'd like

```python
better_graphs.make_better_graph(ax, x_new_ticks={0:0, 50:50, 100:'100m'}, y_new_ticks={0:0, 50:50, 100:'100\%'}, only_new_ticks='xy')
fig
```

## You can customize other elements of the plot at any point

```python
ax.set_xlabel('Distance')
ax.set_ylabel('Proportion')
fig
```
